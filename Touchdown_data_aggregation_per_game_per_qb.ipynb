```python
import pandas as pd

# Assuming 'files' list is defined in another cell and includes file names like:
# files = ['final_filtered_file_2006.csv', 'final_filtered_file_2007.csv', ..., 'final_filtered_file_2022.csv']

all_final_data = pd.DataFrame()

for file in files:
    data = pd.read_csv(file)

    # Calculate total rushing plays for each QB per game
    rushing_plays = data[data['play_type_nfl'] == 'RUSH']
    rush_counts_game = rushing_plays.groupby(['old_game_id', 'qb_id']).size().reset_index(name='qb_rush_plays_per_game')

    # Calculate total touchdowns, pass touchdowns, and rush touchdowns per game
    td_plays = data[data['touchdown'] == 1]
    total_td_game = td_plays.groupby(['old_game_id', 'qb_id']).size().reset_index(name='total_td_per_game')
    pass_td_game = td_plays[td_plays['play_type_nfl'] == 'PASS'].groupby(['old_game_id', 'qb_id']).size().reset_index(name='pass_td_per_game')
    rush_td_game = td_plays[td_plays['play_type_nfl'] == 'RUSH'].groupby(['old_game_id', 'qb_id']).size().reset_index(name='rush_td_per_game')

    # Calculate total rushing plays for each QB per season
    rush_counts_season = rushing_plays.groupby('qb_id').size().reset_index(name='qb_total_rush_plays')

    # Group and aggregate data by old_game_id and qb_id
    grouped_data = data.groupby(['old_game_id', 'qb_id']).agg({
        'defteam': 'first',
        'time_of_day': 'first',
        'temp': 'first',
        'wind': 'first',
        'roof': 'first',
        'surface': 'first'
    }).reset_index()

    # Merge the touchdown counts and rushing counts into the grouped data
    final_data = grouped_data.merge(total_td_game, on=['old_game_id', 'qb_id'], how='left')
    final_data = final_data.merge(pass_td_game, on=['old_game_id', 'qb_id'], how='left')
    final_data = final_data.merge(rush_td_game, on=['old_game_id', 'qb_id'], how='left')
    final_data = final_data.merge(rush_counts_game, on=['old_game_id', 'qb_id'], how='left')
    final_data = final_data.merge(rush_counts_season, on='qb_id', how='left')

    # Fill NaN values with 0
    final_data['qb_rush_plays_per_game'] = final_data['qb_rush_plays_per_game'].fillna(0)
    final_data['qb_total_rush_plays'] = final_data['qb_total_rush_plays'].fillna(0)
    final_data['total_td_per_game'] = final_data['total_td_per_game'].fillna(0)
    final_data['pass_td_per_game'] = final_data['pass_td_per_game'].fillna(0)
    final_data['rush_td_per_game'] = final_data['rush_td_per_game'].fillna(0)

    # Sort data by old_game_id and qb_id
    final_data = final_data.sort_values(by=['old_game_id', 'qb_id'])

    # Concatenate the processed data of each year into the main DataFrame
    all_final_data = pd.concat([all_final_data, final_data], ignore_index=True)

# Save the complete dataset to CSV
all_final_data.to_csv('final_filtered_file_all_qbs2_all_seasons_with_per_game_with_touchdowns.csv', index=False)

# Print summary of the processed data
print("Total Unique Games:", all_final_data['old_game_id'].nunique())
print("Total Rows in Final Dataset:", len(all_final_data))
```
